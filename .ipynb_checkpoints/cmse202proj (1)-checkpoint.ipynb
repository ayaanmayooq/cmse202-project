## Multiple Regression: Ordinary Least Squares (OLS) Model

The Ordinary Least Squares (OLS) model is a regression model that calculates the statistical relationship between an x (independent) and y (dependent) variable. For the purpose of our research question, using the OLS model allows us to establish a relationship between the S&P500 (the x variable) and the different S&P500 stocks such as Google, Tesla, etc. (dependent variables) in order for us to determine how strongly the S&P500 influences stocks.  

We created a function stock_correlation() that takes in the name of a stock and calculates it correlation with the S&P500 using the OLS model. As part of this function, we implemented the fit() and summary() methods in order to fit our data to the line of regression and output statistical details regarding out two variables' correlation. The specific values we will be looking at is the coefficient of determination (R-squared value) in order to understand to what extent the movements of the S&P500 influence the movements of different stocks. 

```python
def stock_correlation(stock_ticker): 
    '''created a function that takes in the stock ticker's name and outputs its correlation to the S&P500'''
    ticker1 = yf.Ticker('SPY') 
    spy_df = ticker1.history(period="24mo") 
    ticker2 = yf.Ticker(stock_ticker)
    stock_df = ticker2.history(period="24mo")
    '''dropping columns that are unnecessary, since we are only calculating the correlation between the closing amounts'''
    x = spy_df.drop(columns=['Open','High','Low','Volume','Dividends','Stock Splits']) # the independent variable is the S&P500
    '''calculating the dependence of the different stocks (y variable) on the S&P500 (x variable)'''
    y = stock_df.drop(columns=['Open','High','Low','Volume','Dividends','Stock Splits']) 

    regression_model = sm.OLS(y,x)
    model_fit = regression_model.fit()
    summary_model = model_fit.summary() #outputs a table with different statistical information regarding the correlation between x and y 

    print(summary_model)
```

## Stock Price Prediction: LSTM

Predicting the price for the GOOGL stock using LSTM.

Loading 10-year stock data.

```python
end = datetime.now()
start = datetime(end.year - 10, end.month, end.day)

data = yf.download('GOOGL', start, end)
data.reset_index(inplace=True)
data
```

```python
data.plot(x='Date', y='Close', figsize=(12, 5))
```

### Training and testing data
We only try to predict the closing price of the stock so we filter out the other columns.

Dividing the satatset into training and testing data.

```python
data = data.filter(['Close', 'Date'])
N = len(data)
train_size = 0.95

train = data.iloc[:int(N*(train_size))]
test = data.iloc[int(N*(train_size)):]
```

```python
plt.plot(train['Close'], label='Training Data')
plt.plot(test['Close'], label='Testing Data')
plt.legend()
plt.show()
```

Min Max scaling the data for easier training.

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(data['Close'].values.reshape(-1,1))
scaled_data
```

Label is the closing prices and the features are the past prices for that stock. The number of days to be considered can be changed.

```python
prediction_days = 60

x_train = []
y_train = []

for i in range(prediction_days, len(train)):
    x_train.append(scaled_data[i - prediction_days: i, 0])
    y_train.append(scaled_data[i, 0])
    
x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

### Building LSTM

Long short-term memory (LSTM) is a recurrent neural network is made up of a collection of artificial neurons. LSTM models are capable of learning long-term dependencies. 

<img src="https://colah.github.io/posts/2015-08-Understanding-LSTMs/img/LSTM3-chain.png" width=500px>

The key to LSTMs is the cell state, the horizontal line running through the top of the diagram.

The cell state is kind of like a conveyor belt. It runs straight down the entire chain, with only some minor linear interactions. Itâ€™s very easy for information to just flow along it unchanged.

The LSTM does have the ability to remove or add information to the cell state, carefully regulated by structures called gates.
Gates are a way to optionally let information through. They are composed out of a sigmoid neural net layer and a pointwise multiplication operation.

The sigmoid layer outputs numbers between zero and one, describing how much of each component should be let through.

Since LSTM models learn from history, it is perfect to use for stock price prediction where we need to consider past stock data to make predictions.

We use **keras** to implement te LSTM model.

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM

model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape= (x_train.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

Our model contains 2 LSTM layers and 2 Dense layers (a deeply connected neural network).

- The first LSTM layer has 128 cells/neurons
- The second LSTM layer has 64 cells/neurons
- The first Dense layer has a 25 neuron deeply connected layer
- The second Dense layer return the prediction

```python
print(model.summary())
```

### Training and Testing the Model

```python
model.fit(x_train, y_train, batch_size=1, epochs=3)
```

```python
test_data = scaled_data[len(train) - 60: , :]

x_test = []
y_test = scaled_data[len(train):, :]
for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])
    
x_test = np.array(x_test)

x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

predictions_scaled = model.predict(x_test)

predictions = scaler.inverse_transform(predictions_scaled)

rmse = np.sqrt(np.mean(((predictions - y_test) ** 2)))
rmse
```

We get a reasonably small Root Mean Square Error (RMSE) meaning that the model has learnt the data well.

## Plotting the Predicitons

```python
pred = test.copy()
pred['Close'] = predictions
```

```python
plt.figure(figsize=(15, 6))
plt.plot(train['Close'])
plt.plot(test['Close'])
plt.plot(pred['Close'])
plt.legend(['Train', 'Test', 'Predictions'])
plt.show()
```

Taking a closer look

```python
plt.figure(figsize=(15, 6))
plt.plot(train['Date'][2200:], train['Close'][2200:])
plt.plot(test['Date'], test['Close'])
plt.plot(pred['Date'], pred['Close'])
plt.legend(['Train', 'Test', 'Predictions'])
plt.show()
```

The prediction is very close to the actual stock price. The model successfully predicts the stock data and overall trend.
